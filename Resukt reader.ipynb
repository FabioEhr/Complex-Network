```python
import pandas as pd

# Read the 2017 country-by-country matrix
df = pd.read_parquet('./dfz_2017_c.parquet')

# For each country (same labels on rows and columns):
#   net_inflow = sum of its column (inflows) – sum of its row (outflows)
net = df.sum(axis=0) - df.sum(axis=1)

# Convert to trillions of dollars
net_trillions = net 

# Sort descending
net_trillions = net_trillions.sort_values(ascending=False)

# Print results
print("Net inflow (column – row) in trillion USD, 2017:\n")
for country, value in net_trillions.items():
    print(f"{country}: {value:.3f} T$")
```
