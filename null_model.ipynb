```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Initialize a list to store DataFrames
dataframes = []

# Load all specified parquet files into the list
for year in range(2010, 2020):
    filename = f"dfz_{year}.parquet"
    df = pd.read_parquet(filename)
    dataframes.append(df)

# Stack all DataFrames to compute mean and standard deviation
data = np.stack([df.values for df in dataframes])

# Compute mean and standard deviation across the loaded DataFrames
mean_df = pd.DataFrame(np.mean(data, axis=0), index=df.index, columns=df.columns)
std_df = pd.DataFrame(np.std(data, axis=0), index=df.index, columns=df.columns)

# Compute percentual standard deviation
percentual_std_df = (std_df / mean_df) * 100

# Plot the percentual standard deviation
plt.figure(figsize=(10, 8))
plt.hist(percentual_std_df.values.flatten(), bins=50, alpha=0.75, color='blue')
plt.title('Distribution of Percentual Standard Deviation of dfz')
plt.xlabel('Percentual Standard Deviation (%)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
